Nous souhaitons maintenant résoudre l'équation de la chaleur 2D en utilisant un schéma d'**Euler explicite** en temps et un schéma aux différences dinies en espace décrit précédemment pour l'[équation de Poisson](poisson.ipynb).

$$
\left\{
\begin{array}{l}
\frac{\partial u}{\partial t}-\Delta u = 0 \; \text{sur} \; \Omega, \\\\
u = 0 \; \text{sur} \; \partial \Omega.
\end{array}
\right.
$$

en prenant la solution initiale suivante

$$
u_0(x, y) = 100e^{-100((x-0.5)^2 + (y -0.5)^2))}.
$$

On prendra comme domaine $\Omega=[0, 1]\times[0,1]$, comme pas d'espace $h_x=0.01$, $h_y=0.01$ et enfin comme pas de temps $dt=h_x^2/4$.

# Mise en place des conditions de bords

```python
def setDirichlet(y):
    y[0, :] = 0
    y[-1, :] = 0
    y[:, 0] = 0
    y[:, -1] = 0
```

# Calcul du terme $\Delta u$

Pour rappel, on calcule de Laplacien par la formule

$$
\Delta u(x_i, y_j) \approx \frac{u_{i-1,j}-2u_{i,j}+u_{i+1,j}}{h_x^2}+\frac{u_{i,j-1}-2u_{i,j}+u_{i,j+1}}{h_y^2}.
$$

```python
def laplacian2D(x, h):
    """
    Produit matrice-vecteur du Laplacien sans assemblage de la matrice

    """
    cx = -1./h[0]**2
    cy = -1./h[1]**2
    c2 = 2./h[0]**2 + 2./h[1]**2 
    y = np.empty(x.shape)
    y[1:-1, 1:-1] = (c2*x[1:-1, 1:-1] 
                     + cy*(x[2:, 1:-1] + x[:-2, 1:-1])
                     + cx*(x[1:-1, 2:] + x[1:-1, :-2]))
    
    return y
```

# Méthode d'Euler explicite dans le cas général

Etant donné un pas de temps $\Delta t$ et une suite d'instants $(t_n = t_0 + n\Delta t)_{n\in \mathbb{N}}$, la méthode d'Euler explicite associée à l'équation différentielle

$$
u'(t) = f(t, u(t))
$$

s'écrit

$$
u_{n+1} = u_n + \Delta t f(t_n, u_n)
$$

```python
def euler(u, dt, f, fargs=()):
    """
    Euler explicite pour un système 

    u'(t) = f(t, u(t))
    
    Parameters :
    ------------

    u: solution à l'instant n
    dt:  pas de temps du schéma
    f: fonction second membre
    fargs: paramètres optionnels de la fonction autre que u

    """
    return u + dt*f(u, *fargs)
```

# Initialisation de la solution

On rappelle que la solution initiale est de la forme

$$
u_0(x, y) = 100e^{-100((x-0.5)^2 + (y -0.5)^2))}.
$$

```python
def initSol(nx, ny, Lx=[0., 1.], Ly=[0., 1.]):
    """
    Initialisation d'une fonction pour tester le schéma en temps

    """
    x = np.linspace(Lx[0], Lx[1], nx)
    y = np.linspace(Ly[0], Ly[1], ny)
    
    x = x[np.newaxis, :]
    y = y[:, np.newaxis]

    return 100.*np.exp(-100.*((x - .5)**2 + (y - .5)**2))
```

# Deuxième version: dérouler les boucles + typage

```python
%%cython -a
import numpy as pnp
    
def laplacian2DCython(double [:, ::1] x, double[::1] h):
    """
    Produit matrice-vecteur du Laplacien sans assemblage de la matrice

    """
    cdef:
        double cx = -1./h[0]**2
        double cy = -1./h[1]**2
        double c2 = 2./h[0]**2 + 2./h[1]**2 
        int nx = x.shape[0]
        int ny = x.shape[1]
        double [:, ::1] y = pnp.empty((nx, ny))
        int i, j
    
    for j in xrange(1, ny-1):
        for i in xrange(1, nx-1):
            y[j, i] = (c2*x[j, i] 
                     + cy*(x[j + 1, i] + x[j - 1, i])
                     + cx*(x[j, i + 1] + x[j, i - 1]))
    
    return y
```

# Troisième version: ajout des directives

```python
%%cython -a 
import numpy as pnp
import cython
    
@cython.boundscheck(False)
@cython.wraparound(False)
@cython.cdivision(True)
def laplacian2DCython(double [:, ::1] x, double[::1] h):
    """
    Produit matrice-vecteur du Laplacien sans assemblage de la matrice

    """
    cdef:
        double cx = -1./h[0]**2
        double cy = -1./h[1]**2
        double c2 = 2./h[0]**2 + 2./h[1]**2 
        int nx = x.shape[0]
        int ny = x.shape[1]
        double [:, ::1] y = pnp.empty((nx, ny))
        int i, j
    
    for j in xrange(1, ny-1):
        for i in xrange(1, nx-1):
            y[j, i] = (c2*x[j, i] 
                     + cy*(x[j + 1, i] + x[j - 1, i])
                     + cx*(x[j, i + 1] + x[j, i - 1]))
    
    return y
```

## Initialisation de la solution initiale en parallèle

On découpe ici notre domaine en rectangles en découpant suivant l'axe $y$.

```python
def initSolPara(nx, ny, Lx=[0., 1.], Ly=[0., 1.]):
    """
    Initialisation d'une fonction pour tester le schéma en temps

    """
    # calcul du pas d'espace en y
    dy = (Ly[1] - Ly[0])/float(ny - 1)

    # Calcul du nombre de points par processus en y sans les points fictifs
    size = mpi.COMM_WORLD.Get_size()
    ny_local = ny/size

    # Calcul des bornes de chaque sous-domaine avec les points fictifs
    rank = mpi.COMM_WORLD.Get_rank()
    deb, fin = Ly[0] + ny_local*rank*dy, Ly[0] + ny_local*(rank + 1)*dy
    if rank > 0: 
        deb -= dy
    if rank == size - 1:
        fin = Ly[1]

    # Calcul du nombre de points par processus en y avec les points fictifs
    nyy = (fin - deb)/dy + 1

    x = np.linspace(Lx[0], Lx[1], nx)
    y = np.linspace(deb, fin, nyy)
    
    x = x[np.newaxis, :]
    y = y[:, np.newaxis]

    return 100.*np.exp(-100.*((x - .5)**2 + (y - .5)**2))
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(initSolPara(101, 101), vmin=0, vmax=100)
plt.show()
```

On doit à présent écrire une fonction permettant de mettre à jour les points ficitifs à chaque itération en temps.

```python
def update_fictitious_point(x):
    rank = mpi.COMM_WORLD.Get_rank()

    down, up = rank - 1, rank + 1
    
    if down < 0:
        down = -1
    if up >= size:
        up = -1

    mpi.COMM_WORLD.Sendrecv([x[1, :], mpi.DOUBLE], down, 0,
                            [x[0, :], mpi.DOUBLE], down, 0)
    mpi.COMM_WORLD.Sendrecv([x[-2, :], mpi.DOUBLE], up, 0,
                            [x[-1, :], mpi.DOUBLE], up, 0)
    
```

Vérifions que tout fonctionne correctement.
